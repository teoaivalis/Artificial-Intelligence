Σε αυτό το σημείο θα εργαστείτε με τα αρχεία ratings τα όποια περιέχουν αξιολογήσεις (από 1 μέχρι 5) για τις παραπάνω ταινίες. Το προηγούμενο σύστημα συστάσεων προτείνει στον χρήστη ταινίες αποκλειστικά με βάση την ομοιότητά τους. Σε αυτό το σημείο θα γίνει μια αναβάθμιση του συστήματος έτσι ώστε να παράγονται καλύτερες  συστάσεις οι όποιες θα λαμβάνουν υπόψιν και τις προτιμήσεις του χρήστη, οι όποιες θα εξάγονται από τις αξιολογήσεις που έχει κάνει μέχρι στιγμής. 

Η εκπαίδευση του recommender θα γίνεται ως εξής:

Για κάθε ταινία θα υπάρχει ένα score το όποιο αρχικά θε είναι ίσο με 0 και θα διαμορφώνεται από τις αξιολογήσεις κάθε user. Έτσι για έναν χρήστη με βάση τις αξιολογήσεις που υπάρχουν στο αρχείο train_ratings θα πρέπει:

1.	Για κάθε ταινία που έχει βαθμολογήσει να βρίσκονται οι κοινές ταινίες ανά κλίμακα και στο μέχρι τώρα σκορ κάθε παρόμοιας ταινίας θα προστίθεται ένα βάρος το όποιο θα μπορούσε να είναι το ποσοστό ομοιότητας της ταινίας (δηλαδή ένα βάρος για κάθε κλίμακα, αν δύο ταινίες μοιάζουν ενισχύουμε το βάρος που προσθέτουμε από το να μοιάζουν λιγότερο) επί τον βαθμό που έχει βάλει ο χρήστης για την αρχική ταινία (διαφορετικό είναι ο χρήστης να έχει βάλει 5/5 ή 1/5 σε μια ταινία από 3/5). 

2. Στη συνέχεια, ανάλογα με το σκορ που έχει σχηματιστεί για κάθε ταινία, θα επιλέγεται αν αυτή θα μπορούσε να είναι προτεινόμενη για τον χρήστη ή όχι και θα μετράμε το πόσο καλά τα πήγε το σύστήμά μας με βάση ορισμένες μετρικές.

Η λογική πίσω από την παραπάνω διαδικασία είναι ότι παρόμοιες ταινίες θα έχουν ανάλογο βαθμό. Για παράδειγμα αν ένας χρήστης έχει αξιολογήσει αρκετές  ταινίες οι όποιες είναι sci-fiction με 5/5 τότε μια ταίνια sci-fiction την όποια δεν έχει δεί λογικά θα του αρέσει και θα έπρεπε να την προτείνουμε.

Μετά την εκπαίδευση του συστήματος σας καλείστε να δοκιμάσετε τον recommender που κατασκευάσατε στην πράξη. Για τον σκοπό αυτό θα φορτώσετε το αρχείο test_ratings.csv όπου περιέχονται οι αξιολογήσεις του ίδιου χρήστη για άλλες ταινίες. Το σύστημά σας πρέπει να προβλέπει αν μια ταινία θα πρέπει να προταθεί στον χρήστη. Μια ταίνια θα έπρεπε να έχει προταθεί στον  χρήστη, αν έχει βαθμό μεγαλύτερο του 3. Συνεπώς για την επίβλεψη του συστήματός σας θα πρέπει για κάθε μια από τις ταινίες του αρχείου test_ratings.csv να επιστρέφεται 1 ή 0 αν η ταινία θα έπαιρνε βαθμό μεγαλύτερο του 3 ή όχι, δηλαδή αν θα έπρεπε να την είχαμε προτείνει ή όχι.

Στην συνέχεια, σε συνδιασμό με τις πραγματικές απαντήσεις του χρήστη θα αξιολογήσετε το σύστημά σας χρησιμοποιόντας τις μετρικές: precision, recall, f1 οι οποίες είναι οι πλέον γνωστές μετρικές και ευρέως χρησιμοποιούμενες τεχνικές για την επίβλεψη-μέτρηση απόδοσης ανάλογων συστημάτων.

1. Precision: Δείχνει πόσο ακριβές είναι το σύστημα. Υπολογίζει πόσα από τα στιγμιότυπα τα όποια προβλέψαμε ότι ανήκουν σε μια κλάση όντως ανήκουν σε αυτή. Η μετρική αυτή μας δίνει μια εικόνα σχετικά με τον αριθμό των ταινίων που προβλέψαμε ως προτεινόμενες ενώ δεν θα έπρεπε.

2. Recall: Υπολογίζει πόσα από τα στιγμιότυπα που ανήκουν σε μια κλάση (π.χ. προτεινόμενες ταινίες) προβλέφθηκαν σωστά.

3. F1: Είναι ένας μέσος μεταξύ των παραπάνω δυο μετρικών, έτσι ώστε να διατηρείται μια ισορροπία μεταξύ τους. Υπολογίζεται από την παρακάτω σχέση:

$$F_1=2\frac{Precision\times{Recall}}{Precision+Recall}$$

Οι παραπάνω συναρτήσεις παρέχονται από την βιβλιοθήκη scikit-learn.

Τέλος για την καλύτερη επίβλεψη του συστήματος σας μπορείτε να εκπαιδεύσετε  τον recommender σας με ένα υποσύνολο ταινίων από ελάχιστες, λίγες μέχρι και πολλές (π.χ. 3, 5, 10, 50, ...) για να μελετήσετε κατά πόσο σας βοηθούν οι επιπλέον αξιολογήσεις κάθε φορά (δηλαδή κατά πόσο βελτιώνονται οι παραπάνω μετρικές στο test set). 

Έτσι π.χ. μπορείτε να εντοπίσετε περιπτώσεις όπως για παράδειγμα ότι με έναν recommender μπορεί να μην επιτυγχάνετε πολύ υψηλό σκορ όσο με άλλους, αλλά το βέλτιστο σκόρ σας επιτυγχάνεται πολύ γρηγόρα π.χ. με μόνο 10 ταινίες αντί 100. Έτσι για παράδειγμα αν ο αλγόριθμος τα πηγαίνει πολύ καλά για τρεις ταινίες και στη συνέχεια το σκορ βελτιώνεται ελάχιστα τότε αυτός ενδείκνυται για ένα σύστημα συστάσεων για νέους χρήστες όπου δεδομένου λίγων ταινιών ο αλγόριθμος είναι σε θέση να το να προτείνει καλές συστάσεις. Ενώ αν η καλύτερη απόδοση του αλγορίθμου σας είναι βέλτιστη με περισσότερες ταινίες π.χ. 50 τότε αυτός ο αλγόριθμος συστάσεων ενδείκνυται για παλιούς χρήστες με πολλές αξιολόγησεις.


# **Μέρος 1: Μελέτη των Metadata, Δημιουργία κόσμου και των βασικών queries.**


````python
import pandas as pd
from pyswip import Prolog
````

````python
# Η βιβλιοθήκη pandas είναι χρήσιμη για την εργασία με τέτοια δεδομένα
import pandas as pd 
# Διάβασμα του αρχείου 'movie_metadata.csv' 
data = pd.read_csv("movies_metadata.csv") 
#Στις θέσεις με nan τιμές βάζουμε 'UNK'
data.fillna("UNK", inplace=True)
# Preview the first 5 lines of the loaded data 
print(data.shape)
print(data.head(0))
data.head()
````

````python
def clean_text(text):
  text = text.replace(u'\xa0', u'')
  text = text.replace(u"'", u'')
  return text
````

Αρχικά για κάθε γραμμή του πίνακα με τις ταινίες φτιάχνουμε τα κατηγορήματα που θελουμε να αποθηκεύσουμε.
Προσέχουμε να μετατρέψουμε όλα τα δεδομένα μας σε string ωστε να προκύψουν οι προτάσεις. 

````python
import ast
#Ορίζουμε τον κόσμο μας
prolog = Prolog()

literals = []
movie_score = {}
for row in data.itertuples(index=True, name='Pandas'):
    budget= clean_text(str(getattr(row,'budget')))
    id= clean_text(str(getattr(row,'id')))
    popularity= clean_text(str(getattr(row,'popularity')))
    gross= clean_text(str(getattr(row,'gross')))
    duration= clean_text(str(getattr(row,'duration')))
    vote_average= clean_text(str(getattr(row,'vote_average')))
    num_voted_users=clean_text(str(getattr(row,'num_voted_users')))
    title_year=clean_text(str(getattr(row,'title_year')))
    movie_title = clean_text(str(getattr(row, 'movie_title')))
    director_name= clean_text(str(getattr(row,'director_name')))
    homepage = clean_text(str(getattr(row,'homepage')))
    language = clean_text(str(getattr(row,'language')))
    original_title = clean_text(str(getattr(row,'original_title')))
    overview = clean_text(str(getattr(row,'overview')))
    status = clean_text(str(getattr(row,'status')))
    tagline = clean_text(str(getattr(row,'tagline')))
    country = clean_text(str(getattr(row,'country')))   
    actor_1_name = clean_text(str(getattr(row,'actor_1_name')))
    actor_2_name = clean_text(str(getattr(row,'actor_2_name')))
    actor_3_name = clean_text(str(getattr(row,'actor_3_name')))
    release_date= clean_text(str(getattr(row,'release_date')))
    

    if movie_title !="UNK":
        if director_name !="UNK":
            literals.append("director('"+ movie_title +"','" + director_name + "')")
        for genre in getattr(row, 'genres').split("|"):
            literals.append("genre('"+ movie_title +"','"+ genre +"')")
        for plot_keyword in clean_text(getattr(row, 'plot_keywords')).split("|"):
            if plot_keyword != "UNK":
                literals.append("plot_keywords('"+ movie_title +"','"+ plot_keyword +"')")

        if budget !="UNK":
            literals.append("budget('"+ movie_title +"'," + budget + ")")
        if id !="UNK":
            literals.append("id('"+ movie_title +"'," + id + ")")
        if popularity !="UNK":
            literals.append("popularity('"+ movie_title +"'," + popularity + ")")
        if gross !="UNK":
            literals.append("gross('"+ movie_title +"'," + gross + ")")
        if duration !="UNK":
            literals.append("duration('"+ movie_title +"'," + duration + ")")
        if vote_average !="UNK":
            literals.append("vote_average('"+ movie_title +"'," + vote_average + ")")
        if num_voted_users !="UNK":
            literals.append("num_voted_users('"+ movie_title +"'," + num_voted_users + ")")
        if title_year !="UNK":
            literals.append("title_year('"+ movie_title +"'," + title_year + ")")
        
        if homepage !="UNK":
            literals.append("homepage('"+ movie_title +"','" + homepage + "')")
        if language !="UNK":
            literals.append("language('"+ movie_title +"','" + language + "')")
        if original_title !="UNK":
            literals.append("original_title('"+ movie_title +"','" + original_title + "')")
        if overview !="UNK":
            literals.append("overview('"+ movie_title +"','" + overview + "')")
        if status !="UNK":
            literals.append("status('"+ movie_title +"','" + status + "')")
        if tagline !="UNK":
            literals.append("tagline('"+ movie_title +"','" + tagline + "')")
        if country !="UNK":
            literals.append("country('"+ movie_title +"','" + country + "')")
        if release_date!="UNK":
            literals.append("release_date('"+ movie_title +"','" + release_date + "')")
        if actor_1_name !="UNK":
            literals.append("actor_1_name('"+ movie_title +"','" + actor_1_name + "')")            
        if actor_2_name !="UNK":
            literals.append("actor_2_name('"+ movie_title +"','" + actor_2_name + "')")              
        if actor_3_name !="UNK":
            literals.append("actor_3_name('"+ movie_title +"','" + actor_3_name + "')")              
    
        l=ast.literal_eval(getattr(row, 'production_companies')) #convert string to list of dictionaries
        if len(l)!=0:
          for productions_companies in l:
            literals.append("production_companies('"+ movie_title +"','" + clean_text(productions_companies['name']) +  "','" + clean_text(str(productions_companies['id'])) + "')") 

        l=ast.literal_eval(getattr(row, 'production_countries')) 
        if len(l)!=0:
          for productions_countries in l:
            literals.append("production_countries('"+ movie_title +"','" + clean_text(str(productions_countries['iso_3166_1'])) +  "','" + clean_text(productions_countries['name']) + "')") 

        l=ast.literal_eval(getattr(row, 'spoken_languages'))
        if len(l)!=0:
          for spoken_languages in l:
            literals.append("spoken_languages('"+ movie_title +"','" + clean_text(str(spoken_languages['iso_639_1'])) +  "','" + clean_text(spoken_languages['name']) + "')") 
                          
                      
literals.sort()            
for literal in literals:
  prolog.assertz(literal)
````

# **2ο Μέρος: Recommendation System με βάση μόνο τα χαρακτηριστικά των ταινιών.**

Στο σημείο αυτό με βάση τους κανόνες που κατασκευάστηκαν στο Μέρος 1 θα κατασκευαστούν κατηγορήματα για την ομοιότητα ταινιών. Παρακάτω δίνεται ένα μικρό παράδειγμα ενός κανόνα και πώς αυτός θα μπορούσε να γραφτεί μέσω του Pyswip. Επίσης όπως αναφέρεται και σε σχόλιο παραπάνω θα μπορεί να γραφτεί και μια βάση δεδομένων με τους κανόνες και να γίνει απευθείας consult.

Στο παρακάτω παράδειγμα το 5 και το 4 εκφράζουν την ομοιότητα των ταινιών π.χ. οι ταινίες που παράγονται μέσω του find_similar_5 είναι πιο όμοιες από αυτές που παράγονται μέσω του find_similar_4. 


Στο σημείο αυτό κατασκευάσαμε queries τα οποία μας επιστρέφουν παρόμοιες ταινίες, με την βοήθεια αυτών που κάναμε στο Μέρος 1. Έχουμε 7 κατηγορήματα, εκ τω των οποίων οι ταινίες που εισάγονται ως παράμετροι στο find_similar_movies_7 έχουν περισσότερα κοινά στοιχεία από αυτές στο find_similar_movies_1 τα λιγότερα. Αυτά είναι:

-find_similar_movies_1 : Αυτές οι ταινίες έχουν σχετικά ίδιο θέμα (μια λέξη κενή) και σχετικά ίδια πλοκή (μια λέξη κοινή).

-find_similar_movies_2 : Αυτές οι ταινίες έχουν ίδια γλώσσα, σχετικά ίδιο θέμα (μια λέξη κενή) και σχετικά ίδια πλοκή (μια λέξη κοινή).

-find_similar_movies_3 : Αυτές οι ταινίες έχουν ίδια γλώσσα, σχετικά ίδιο θέμα (μια λέξη κενή) και αρκετή ίδια πλοκή (δύο λέξεις κοινές).

-find_similar_movies_4 : Αυτές οι ταινίες έχουν ίδια γλώσσα, αρκετά ίδιο θέμα (δύο λέξεις κενές) και αρκετά ίδια πλοκή (δύο λέξεις κοινές).

-find_similar_movies_5 : Αυτές οι ταινίες έχουν έναν κοινό ηθοποιό, ίδια γλώσσα, αρκετά ίδιο θέμα (δύο λέξεις κενές) και πολύ ίδια πλοκή (τρεις λέξεις κοινές).

-find_similar_movies_6 : Αυτές οι ταινίες έχουν δύο κοινούς ηθοποιούς, ίδια γλώσσα, αρκετά ίδιο θέμα (δύο λέξεις κενές) και πολύ ίδια πλοκή (τρεις λέξεις κοινές).

-find_similar_movies_7 : Αυτές οι ταινίες έχουν ίδιο σκηνοθέτη, τρεις κοινούς ηθοποιούς, ίδια γλώσσα, πολύ ίδιο θέμα (τρεις λέξεις κενές) και πολύ ίδια πλοκή (τρεις λέξεις κοινές).

Σε αυτό το μέρος θα πάρουμε να βρούμε τις προτεινόμενες ταινίες για κάθε ένα από τα find_similar που έχουμε φτιάξει. Θα συνενώσουμε όλες τις λίστες που προκύπτουν και το αποτέλεσμα θα είναι το recomendation μας προς τον χρήστη.

````python
movie_dict={}

def simple_recommender(movie):
    movie_dict.clear()
    s7 = set()
    q = prolog.query("find_similar_movies_7('" + movie +"',M)")    
    for soln in q:
        m = soln['M'] 
        if m not in s7:
            s7.add(soln['M'])
            movie_dict[m]=7
    q.close()

    s6 = set()
    q = prolog.query("find_similar_movies_6('" + movie +"',M)")
    for soln in q:
        m = soln['M'] 
        if m not in s6 and m not in s7:
            s6.add(soln['M'])
            movie_dict[m]=6
    q.close()

    s5 = set()
    q = prolog.query("find_similar_movies_5('" + movie +"',M)")
    for soln in q:
        m = soln['M'] 
        if m not in s5 and not m in s6 and not m in s7:
            s5.add(soln['M'])
            movie_dict[m]=5
    q.close()

    s4 = set()
    q = prolog.query("find_similar_movies_4('" + movie +"',M)")
    for soln in q:
        m = soln['M'] 
        if m not in s4 and m not in s5 and not m in s6 and not m in s7:
            s4.add(soln['M'])
            movie_dict[m]=4
    q.close()

    s3 = set()
    q = prolog.query("find_similar_movies_3('" + movie +"',M)")
    for soln in q:
        m = soln['M'] 
        if m not in s3 and m not in s4 and m not in s5 and not m in s6 and not m in s7:
            s3.add(soln['M'])
            movie_dict[m]=3
    q.close()

    s2 = set()
    q = prolog.query("find_similar_movies_2('" + movie +"',M)")
    for soln in q:
        m = soln['M'] 
        if m not in s2 and m not in s3 and m not in s4 and m not in s5 and not m in s6 and not m in s7:
            s2.add(soln['M'])
            movie_dict[m]=2
    q.close()

    s1 = set()
    q = prolog.query("find_similar_movies_1('" + movie +"',M)")
    for soln in q:
        m = soln['M'] 
        if  m not in s1 and m not in s2 and m not in s3 and m not in s4 and m not in s5 and not m in s6 and not m in s7:
            s1.add(soln['M'])
            movie_dict[m]=1
    q.close()
    return list(sorted(s7))+list(sorted(s6))+list(sorted(s5))+list(sorted(s4))+list(sorted(s3))+list(sorted(s2))+list(sorted(s1))
````

# **3ο Μέρος: Recommendation System Με βάση τις προτιμήσεις-Αξιολογήσεις του χρήστη-Εκπαίδευση και Πρόβλεψη**


Αρχικά μελετάμε τις αξιολογήσεις κάθε χρήστη για να καταλάβουμε την δομή και τις πληροφορίες κάθε αρχείου.




Κατασκευάζουμε τα scores για κάθε ταινία με βάση τις προηγούμενες αξιολογήσεις του χρήστη

Στο σημείο αυτό κατασκευάζουμε την νέα συνάρτηση η όποια θα επιστρέφει τις προτεινόμενες ταινίες συνυπολογίζοντας τις αξιολογήσεις του χρήστη.

Αρχικά ορίζουμε τα rating_weights και score_weights. 

Τα rating_weights ανατέθηκαν συμμετρικά αρνητικές και θετικές τιμές σε όλες τις αξιολογήσεις απο 0 έως 5 ώστε να επηρεάζουν σε κάποιο βαθμό τις αντίστοιχες αξιολογήσεις άλλων ταινιών όλα τα ratings.

Τα score_weights τέθηκαν ανάλογα με τα επίπεδα ομοιότητας που ορίστηκαν στην simple_recommender με μικρές διαφοροποιήσεις που εξαρτώνται από το εύρος των αλλαγών που έγιναν από το ένα επίπεδο στο άλλο και το βαθμό που αυτές τα διαφοροποιούν. Συνεπώς αν ανάμεσα σε 2 επίπεδα προστέθηκε ένα λιγότερο σημαντικό χαρακτηριστικό αυξάνουμε το score_weight κατά 0.5 ενώ για αλλαγή σημαντικότερου χαρακτηριστικού 1.

Για την υλοποίσηση της predict έχουμε ορίσει ότι μια ταινία θα έπρεπε να είναι προτεινόμενη αν είχε σκορ > 3. Και βάζουμε ότι μια ταινία είναι προτεινόμενη αν έχει σκορ > 0.

````python
from tqdm.notebook import tqdm
from sklearn.metrics import precision_score, recall_score, f1_score
import numpy as np
import random

rating_weights = {0: -2, 1: -1, 2:-0.5, 3:0.5, 4:1, 5:2}
score_weights = {1: 1, 2: 1.5, 3: 2.5, 4: 3.5, 5: 5, 6: 6.5, 7: 8} # ανάλογα με τα επίπεδα ομοιότητας που έχουν οριστεί στην simple_recommender
movie_score={}

def train_recommender(ratings, rating_weights, score_weights, number_of_movies = 10):
    """
    Στην συνάρτηση αυτή μπορούμε να ορίζουμε ποιο υποσύνολο των αξιολογήσεων θα χρησιμοποιήσουμε για το train μαζί με τα βάρη ομοιότητας και σκορ
    Σε συνδυασμό με τον αριθμό των ταινιών που θέλουμε να χρησιμοποιήσουμε σαν σύνολο δεδομένων π.χ. 10 από τις 100 ή 3 από τις 100 κ.ο.κ
    Αν θέλουμε να χρησιμοποιήσουμε όλες τις ταινίες σαν training set τότε ορίζουμε το number_of_movies = - 1
    """
    if number_of_movies > len(ratings):
        number_of_movies = len(ratings)

    if number_of_movies != -1:
        indexes = random.sample(range(len(ratings)), number_of_movies)
        ratings = ratings.iloc[indexes]


    for row in tqdm(ratings.itertuples(index=True, name='Pandas')):
        movie = clean_text(getattr(row, 'movie_title'))
        rating = getattr(row, 'rating')

        similar_movies = simple_recommender(movie) # Βρίσκουμε όλες τις ταινίες που προτείνονται για τη συγκεκριμένη

        for similar_movie in similar_movies:  # Για κάθε τέτοια ταινία που δεν την έχει δεί ήδη ο χρήστης (άρα βαθμολογήσει)
            if similar_movie not in movie_score:
                movie_score[similar_movie] = rating_weights[int(rating)] * score_weights[movie_dict[similar_movie]]               
            else:
                movie_score[similar_movie] += rating_weights[int(rating)] * score_weights[movie_dict[similar_movie]] #το weight θα το ορίσετε ανα επίπεδο ομοιότητας οι πολύ όμοιες ταινίες θα έχουν μεγαλύτερο βάρος

    return movie_score


def predict_example(ratings, movie_score):
    real, pred = [], []
    for i, row in enumerate(ratings.itertuples(index=True, name='Pandas')):
        movie = clean_text(getattr(row, 'movie_title'))
        rating = getattr(row, 'rating')

        if movie in movie_score: #αν έχουμε σχηματίσει βαθμολογία για την ταινία αυτή
            pred.append(int(movie_score[movie] > 0)) #heuristic για το αν μια ταινία είναι προτεινόμενη
            real.append(int(rating > 3))# έτσι ορίζουμε ότι μια ταινία θα έπρεπε να είναι προτεινόμενη
            #η συνθήκη αυτή δεν μπορεί να αλλάξει
        else: #δεν μπορούμε να προτείνουμε κάτι για το όποιο δεν έχουμε σχηματίσει εικόνα
            pred.append(0)
            real.append(int(rating > 3))

    return real, pred


def get_metrics(real, pred):
    metrics = {}
    metrics["precision"] = precision_score(real, pred)
    metrics["recall"] = recall_score(real, pred)
    metrics["f1"] = f1_score(real, pred)
    return metrics
````
